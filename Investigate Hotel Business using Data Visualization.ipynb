## Total durasi menginap

```python
data3['total_duration_day'] = data3['stays_in_weekend_nights'] + data3['stays_in_weekdays_nights']
data3['total_duration_day'].describe()
```

```python
plt.figure(figsize=(12,8))
plt.hist(data3['total_duration_day'], bins=20)
```

## Pengelompokan type visitor berdasarkan durasi menginap

```python
data3['Booking Type'] = np.where(data3['total_duration_day']<= 3, 'Short', 
                         (np.where(data3['total_duration_day']< 10, 'Medium', 'Long')))
```

Alasan pemilihan hari karena rata-rata pengunjung menginap selama 3 hari, 10 hari titik puncak maksimal pada histogram, dan selebihnya di kategorikan long karena menginap lebih dari 10 malam. 

```python
data3['Booking Type'].describe()
```

## Perbandingan jumlah pemesanan hotel yang dibatalkan terhadap durasi menginap

```python
data3_cancel = data3.groupby(['is_canceled','hotel']).agg({'total_duration_day':['sum', 'count']}).reset_index()
data3_cancel
```

```python
hotel_canceled = data3.groupby(['hotel','Booking Type']).agg({'is_canceled':'sum'}).reset_index()
hotel_canceled
```

```python
total_cancel = hotel_canceled.groupby(['hotel']).agg({'is_canceled':'sum'}).reset_index()
total_cancel #Total cancel secara keseluruhan untuk semua hotel
```

```python
hotel_cancel = hotel_canceled.merge(total_cancel, on = 'hotel', how='left')
hotel_cancel
```

```python
hotel_cancel['presentage_cancel'] = hotel_cancel['is_canceled_x']/hotel_cancel['is_canceled_y']
kolom_cancel = ['hotel','Booking_Type', 'total_reservasi', 'total_rsvp_cancel', 'ratio_cancel']
hotel_cancel.columns = kolom_cancel
hotel_cancel
```

Booking Type based on duration :

```python
data4['Cat_lead_Time'] = np.where(data4['lead_time']<= 80, 'Short', 
                         (np.where(data4['lead_time']<200, 'Medium', 'Long')))
```

```python
data4['Cat_lead_Time'].describe()
```

## Perbandingan jumlah pemesanan hotel yang dibatalkan terhadap jarak waktu pemesanan

```python
cancel_lead = data4.groupby(['hotel','Cat_lead_Time']).agg({'is_canceled':'sum'}).reset_index()
cancel_lead
```

```python
total_cancel = data4.groupby(['hotel']).agg({'is_canceled':'sum'}).reset_index()
total_cancel
```

```python
canceled_lead = cancel_lead.merge(total_cancel, on = 'hotel', how='left')
canceled_lead
```

```python
canceled_lead.columns
```

```python
kolom_canceled_lead=['hotel', 'Cat_lead_Time',  'cancel_lead', 'total_cancel']
canceled_lead.columns = kolom_canceled_lead
```

```python
canceled_lead['ratio_cancel'] = canceled_lead['cancel_lead'] / canceled_lead['total_cancel']
canceled_lead
```

## Visualisasi rasio pembatalan pesanan terhadap jarak waktu pemesanan

```python
plt.figure(figsize=(12,8))
ax2 = sns.barplot(x='Cat_lead_Time', y='ratio_cancel', data= canceled_lead, hue='hotel', palette='Paired')
plt.title('Rasio pembatalan pesanan terhadap jarak waktu pemesanan\n\n\n\n', fontsize = 18,fontweight ='bold', color='darkblue', loc='center')
plt.title('Secara umum kedua tabel memiliki tingkat pembatan identik untuk setiap kategori. \nPada kategori short yaitu kurang dari 85 hari memiliki tingkat pembatalan yang tinggi dibandingkan orang yang\nmelakukan reservasi jauh-jauh hari. Tingkat pembatalan orang yang melakukan reservasi pada type perkotaan \n52%, dan pada type resort 46%', fontsize = 12, loc ='left', color='black')

    
```

## Interpretasi :
Secara umum kedua tabel memiliki tingkat identik untuk setiap kategori. Pada kategori short yaitu kurang dari 85 hari memiliki tingkat pembatalan yang tinggi dibandingkan orang yang melakukan reservasi jauh-jauh hari. Tingkat pembatalan orang yang melakukan reservasi pada type perkotaan 52%, dan pada type resort 46%. Dengan begini, dapat di berikan rekomendasi nominal booking pad hotel untuk setiap booking sebelum 85 hari memiliki pembayaran yang lebih tinggi daripada kategori medium ataupun long. Hal ini dilakukan untuk mengurangi angka pembatalan pada setiap reservasi.

Insight :
Tingkat pembatalan secara berpola dapat dilihat semakin lama waktu reservasi dari hari pemesanan, maka semakin berpotensi untuk melakukan pemesanan. Hal itu ditunjukkan dari rendahnya tingkat pembatalan.

Rekomendasi yang dapat diberikan untuk kedua jenis type hotel adalah penetapan harga untuk melakukan reservasi berdasarkan waktu reservasi.

Urutan berdasarkan jarak waktu pemesanan dari yang termahal ke termurah yaitu :
